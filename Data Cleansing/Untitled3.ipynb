```python
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch stock prices for AAPL and NVDA
aapl = yf.Ticker('AAPL')
NVDA = yf.Ticker('NVDA')

# Get historical market data
hist_aapl = aapl.history(period='1y')
hist_NVDA = NVDA.history(period='1y')

# Normalize the data
hist_aapl = hist_aapl['Close'] / hist_aapl['Close'].iloc[0]
hist_NVDA = hist_NVDA['Close'] / hist_NVDA['Close'].iloc[0]
```

```python
import pandas as pd
# Create dataframe for easy plotting
stocks = pd.DataFrame({
    'AAPL': hist_aapl,
    'NVDA': hist_NVDA
})
```

```python
# Plot data
stocks.plot(grid = True)
plt.title('Normalized stock prices of AAPL and NVDA over the last year')
plt.show()
```
