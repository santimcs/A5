```python
sql = """
SELECT P.name, T.id
FROM profits P
JOIN tickers T
ON P.name = T.name
ORDER BY P.name
"""
profits = pd.read_sql(sql, conpg)
profits.shape
```

```python
sql = """
SELECT P.name, S.price, inc_pct_y, inc_pct_q, inc_pct_py, inc_pct_pq, mean_pct, std_pct
FROM profits P
JOIN tickers T
ON P.ticker_id = T.id
LEFT JOIN stocks S
ON S.ticker_id = T.id
ORDER BY P.name
"""
profits = pd.read_sql(sql, conpg)
profits.shape
```

```python
sql = """
SELECT P.name, S.price, inc_pct_y, inc_pct_q, inc_pct_py, inc_pct_pq, mean_pct, std_pct
FROM profits P
JOIN stocks S
ON P.ticker_id = S.ticker_id
ORDER BY P.name
"""
profits = pd.read_sql(sql, conpg)
profits.shape
```

```python
sql = """
SELECT P.name, S.price, inc_pct_y, inc_pct_q, inc_pct_py, inc_pct_pq, mean_pct, std_pct
FROM profits P
LEFT JOIN stocks S
ON P.ticker_id = S.ticker_id
WHERE S.id IS null
ORDER BY P.name
"""
profits = pd.read_sql(sql, conpg)
profits
```

```python
sql = '''
SELECT name, market
FROM stocks
ORDER BY name'''
my_stocks = pd.read_sql(sql, conpg)
my_stocks.dtypes
```

```python
filters = [
   (my_stocks.market.str.contains('SET50')),
   (my_stocks.market.str.contains('SET100')),
   (my_stocks.market.str.contains('mai'))    
]
values = ['SET50','SET100','mai']
```

```python
my_stocks["mrkt"] = np.select(filters, values, default='SET999')
my_stocks.head()
```

```python
df_merge = pd.merge(profits, my_stocks, on='name', how='inner')
df_merge.groupby('mrkt').count()
```

```python
set50 = df_merge[df_merge['mrkt'] == 'SET50']
set50.sort_values('name',ascending=True)
```

```python
set100 = df_merge[df_merge['mrkt'] == 'SET100']
set100.sort_values('name',ascending=True)
```

```python
set999 = df_merge[df_merge['mrkt'] == 'set999']
set999.sort_values('name',ascending=True)
```
