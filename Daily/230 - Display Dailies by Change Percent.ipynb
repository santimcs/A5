```python
#df_dailies = pd.read_csv('../data/dailies.csv')
df_dailies = pd.read_csv('../data/Price-Today.csv')
#df_dailies.rename(columns={'daily_volume':'value(m)'},inplace=True)
df_dailies.columns
```

```python
#df_dailies.loc[(df_dailies['value(m)'] == '-') ,df_dailies['value(m)']] = 0

df_dailies['value(m)'] = df_dailies['value(m)'].astype(float)
df_dailies['value(m)'] = round(df_dailies['value(m)'] / 1000,0)
df_dailies['value(m)'] = df_dailies['value(m)'].astype(int)
df_dailies.dtypes
```
