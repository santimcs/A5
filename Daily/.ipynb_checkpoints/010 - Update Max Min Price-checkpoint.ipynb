```python
import os
import pandas as pd
from datetime import date, timedelta
from sqlalchemy import create_engine
engine = create_engine(
    "postgresql+psycopg2://postgres:admin@localhost:5432/portpg_development"
)
conpg = engine.connect()
```

```python
sql = '''
SELECT *
FROM stocks
'''
stocks = pd.read_sql(sql, conpg)
stocks.dtypes
```

```python
input_file = 'price-hilo.csv'
current_dir = '../Weekly'
input_dir = os.path.join(os.path.dirname(current_dir), 'Data')
# output_dir = os.path.join(os.path.dirname(current_dir), 'Data')
print(input_dir)
```

```python
file_in = os.path.join(input_dir, input_file)
# file_out = os.path.join(output_dir, "price-hilo.csv")
print(file_in)
```

```python
in_df = pd.read_csv(file_in)
in_df.dtypes
```

```python
in_df
```

### Update table Stocks in PortPG Database, columns max_price & min_price with 
### year_high & year_low from price-hilo.csv

```python
for index, row in in_df.iterrows():
    sql = '''
        UPDATE stocks
        SET price = %s, max_price = %s, min_price = %s, updated_at = %s WHERE name = %s
    '''
    print(sql % (row['price'], row['year_high'], row['year_low'], row['updated_at'], row['name']))
    conpg.execute(sql, (row['price'], row['year_high'], row['year_low'], row['updated_at'], row['name']))
```
