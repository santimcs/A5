```python
import csv
import os
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
from time import sleep, strftime

input_file = 'name-tmp.csv'
current_dir = '../Data Cleansing'
input_dir = os.path.join(os.path.dirname(current_dir), 'Data')
output_dir = os.path.join(os.path.dirname(current_dir), 'Data')
file_in = os.path.join(input_dir, input_file)
file_out = os.path.join(output_dir, "price-hilo-df.csv")
fo = open(file_out, "w")
fi = open(file_in, "r")
print(file_in, file_out)
```

```python
header = 'name, today_low, today_high, year_low, year_high'
url = "https://www.set.or.th/th/market/product/stock/quote/"

start_time = strftime("%I:%M %p")
print('Start at: ', start_time)
```

```python
with open(file_in, newline='', mode='r') as csv_file:
    reader = csv.reader(csv_file)
#     next(reader)  # Skip header row
    out_line = 'name,today_low,today_high,year_low,year_high'
    out_line += '\n'  
    fo.write(out_line)
    for row in reader:
        array = []
        name = row[0].upper()

        response = urlopen(url + name + '/price')
        html_page = soup(response, 'html.parser')
        array.append(name)
        
        values_low = html_page.find_all('span', class_='title-font-family fs-16px fw-bolder me-auto lh-1')
        today_low = values_low[0].text
        if (today_low == '-'):
            today_low = '0.00'
        array.append(today_low)
             
        values_high = html_page.find_all('span', class_='title-font-family fs-16px fw-bolder lh-1')
        today_high = values_high[0].text
        if (today_high == '-'):
            today_high = '0.00'        
        array.append(today_high)        
        
        year_low = values_low[1].text       
        array.append(year_low)
        
        year_high = values_high[1].text       
        array.append(year_high)       
            
        out_line = ','.join(array)
        print(out_line)         
        out_line += '\n'       
        fo.write(out_line)

        sleep(1)
        
fo.close()        
```

```python
end_time = strftime("%I:%M %p")
print('End at: ', end_time)
```

```python
import os
import pandas as pd
from urllib import request
from bs4 import BeautifulSoup
url = "https://www.set.or.th/en/market/product/stock/quote/"

input_file = 'name-tmp.csv'
current_dir = '../Daily'
input_dir = os.path.join(os.path.dirname(current_dir), 'Data')
output_dir = os.path.join(os.path.dirname(current_dir), 'Data')
file_in = os.path.join(input_dir, input_file)
file_out = os.path.join(output_dir, "price-hilo-df.csv")
print(file_in, file_out)
```

```python
# output_columns = ['name', 'today_low', 'today_high', 'year_low', 'year_high']
# output_df = pd.DataFrame(columns=output_columns) 
# output_df.dtypes
```

```python
def get_stock_data(input_df):
    # Assuming the first row is header, use `usecols` parameter of read_csv method to skip it.
    input_df = pd.read_csv(file_in, header=0, usecols=['Name'])
    
    output_columns = ['name', 'today_low', 'today_high', 'year_low', 'year_high']
    output_df = pd.DataFrame(columns=output_columns)  # create an empty dataframe with required columns
    
    for index, row in input_df.iterrows():
        name = str(row['Name']).upper()
        response = request.urlopen(url + name + '/price')
        html_page = BeautifulSoup(response, 'html.parser')
        
        values_low = html_page.find_all('span', class_='title-font-family fs-16px fw-bolder me-auto lh-1')
        today_low = values_low[0].text
        if (today_low == '-'):
            today_low = '0.00'
        
        values_high = html_page.find_all('span', class_='title-font-family fs-16px fw-bolder lh-1')
        today_high = values_high[0].text
        if (today_high == '-'):
            today_high = '0.00'         
        
        year_low = values_low[1].text        
        year_high = values_high[1].text        
#         output_row = pd.DataFrame([(name, today_low, today_high, year_low, year_high)], index=output_columns)
        output_row = pd.DataFrame([name, today_low, today_high, year_low, year_high])        
        
        # append new row to the output dataframe
        output_df = output_df.append(output_row, ignore_index=True)
#         sleep(1)   # to avoid overloading the server and getting blocked
    
    return output_df  # return the final dataframe with all values populated
```

```python
input_df = pd.read_csv(file_in, header=0, usecols=['Name'])
output_df = get_stock_data(input_df)
output_df.dtypes
```
