```python
import csv
import os
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
from time import sleep, strftime

input_file = 'name-ttl.csv'
current_dir = '../Data Cleansing'
input_dir = os.path.join(os.path.dirname(current_dir), 'Data')
output_dir = os.path.join(os.path.dirname(current_dir), 'Data')
file_in = os.path.join(input_dir, input_file)
file_out = os.path.join(output_dir, "price-hilo.csv")
fo = open(file_out, "w")
fi = open(file_in, "r")
print(file_in, file_out)
```

```python
header = 'name, today_low, today_high, year_low, year_high'
url = "https://www.set.or.th/th/market/product/stock/quote/"

start_time = strftime("%I:%M %p")
print('Start at: ', start_time)
```

```python
with open(file_in, newline='', mode='r') as csv_file:
    reader = csv.reader(csv_file)
#     next(reader)  # Skip header row
    out_line = 'name,today_low,today_high,year_low,year_high'
    out_line += '\n'  
    fo.write(out_line)
    for row in reader:
        array = []
        name = row[0].upper()

        response = urlopen(url + name + '/price')
        html_page = soup(response, 'html.parser')
        array.append(name)
        
        values_low = html_page.find_all('span', class_='title-font-family fs-16px fw-bolder me-auto lh-1')
        today_low = values_low[0].text
        if (today_low == '-'):
            today_low = '0.00'
        array.append(today_low)
             
        values_high = html_page.find_all('span', class_='title-font-family fs-16px fw-bolder lh-1')
        today_high = values_high[0].text
        if (today_high == '-'):
            today_high = '0.00'        
        array.append(today_high)        
        
        year_low = values_low[1].text       
        array.append(year_low)
        
        year_high = values_high[1].text       
        array.append(year_high)       
            
        out_line = ','.join(array)
        print(out_line)         
        out_line += '\n'       
        fo.write(out_line)

        sleep(1)
        
fo.close()        
```

```python
end_time = strftime("%I:%M %p")
print('End at: ', end_time)
```
